```python
with open('easy.txt', 'r') as f:
    lines = f.read().splitlines()
folds = lines[-2:]
dots = lines[:-3]
```

```python
import numpy as np
max_x, max_y = 0, 0
for fold in folds:
    axis, size = fold.split(' ')[-1].split('=')
    if axis == 'x' and int(size) > max_x:
        max_x = int(size)
    if axis == 'y' and int(size) > max_y:
        max_y = int(size)
    
paper = np.full((max_y * 2 + 1, max_x * 2 + 1), 0)

for dot in dots:
    x, y = map(lambda x: int(x), dot.split(','))
    paper[y][x] = 1
    
def fold_paper(paper, axis, line):
    if axis == 'x':
        left = paper[:, :line]
        right = paper[:, line+1:]
        mirror_right = np.flip(right, 1)
        left[:, -right.shape[1]:] += mirror_right
        return left
    else:
        up = paper[:line, :]
        down = paper[line+1:, :]
        mirror_down = np.flip(down, 0)
        up[-down.shape[0]:, :] += mirror_down
        return up
        
print(paper)
for fold in folds:
    axis, line = fold.split(' ')[-1].split('=')
    paper = fold_paper(paper, axis, int(line))
    print(len(paper[np.where(paper > 0)]))
    print(paper)
```

```python
print('ol')
for i, row in enumerate(paper):
    row = ''
    for j, value in enumerate(paper[i]):
        if paper[i][j] > 0:
            row += '#'
        else:
            row += '.'
    print(row)
print('ll')
```
