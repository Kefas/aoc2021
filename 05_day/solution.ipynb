```python
with open('input.txt', 'r') as f:
    inputs = f.read().splitlines()

max_x = 0
max_y = 0

lines = []

for inp in inputs:
    x1y1, x2y2 = inp.split('->')
    x1, y1 = x1y1.split(',')
    x2, y2 = x2y2.split(',')
    
    lines.append(((int(x1), int(y1)), (int(x2), int(y2))))
    
                 
for (x1, y1), (x2, y2) in lines:
    if x1 > max_x:
        max_x = x1
    elif x2 > max_x:
        max_x = x2
    
    if y1 > max_y:
        max_y = y1
    elif y2 > max_y:
        max_y = y2
        
diagram = []
for i in range(max_x + 1):
    row = []
    for j in range(max_y + 1):
        row.append(0)
    diagram.append(row)
    
```

```python
def fill(diagram, x1, y1, x2, y2):
    if x1 == x2:
        print(x1, y1, x2, y2)
        if y1 < y2:
            for j in range(y1, y2 + 1):
                diagram[x1][j] += 1
        else:
             for j in range(y2, y1 + 1):
                diagram[x1][j] += 1
    elif y1 == y2:
        print(x1, y1, x2, y2)
        if x1 < x2:
            for i in range(x1, x2 + 1):
                diagram[i][y1] += 1
        else:
            for i in range(x2, x1 + 1):
                diagram[i][y1] += 1

for (x1, y1), (x2, y2) in lines:
    fill(diagram, x1, y1, x2, y2)
```

```python
result = 0
for row in diagram:
    for value in row:
        if value > 1:
            result += 1
```

```python
print(result)
```

```python
def fill_2(diagram, x1, y1, x2, y2):
    if x1 == x2:
        if y1 < y2:
            for j in range(y1, y2 + 1):
                diagram[x1][j] += 1
        else:
             for j in range(y2, y1 + 1):
                diagram[x1][j] += 1
    elif y1 == y2:
        if x1 < x2:
            for i in range(x1, x2 + 1):
                diagram[i][y1] += 1
        else:
            for i in range(x2, x1 + 1):
                diagram[i][y1] += 1
    else:
#         print(f'({x1}, {y1}) -> ({x2}, {y2})')
        if x1 < x2:
            change_x = 1
        else:
            change_x = -1
        
        if y1 < y2:
            change_y = 1
        else:
            change_y = -1
            
        diff = abs(x1 - x2)
        
        for i in range(diff + 1):
#             print(f'x1 + {i * change_x} y1 + {i * change_y}')
            diagram[x1 + i * change_x][y1 + i * change_y] += 1

for (x1, y1), (x2, y2) in lines:
    fill_2(diagram, x1, y1, x2, y2)

```

```python
result = 0
for row in diagram:
    for value in row:
        if value > 1:
            result += 1

```

```python
print(result)
```
