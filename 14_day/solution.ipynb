```python
with open('easy.txt', 'r') as f:
    lines = f.read().splitlines()
```

```python
template = lines[0]
instructions = lines[2:]

instructions_dict = {}
for instruction in instructions:
    key, value = instruction.split(' -> ')
    instructions_dict[key] = value
```

```python
steps = 10
polymer = template
for step in range(steps):
    products = ''
    for i in range(len(polymer)-1):
        products += instructions_dict[polymer[i:i+2]]
    
    new_polymer = ''
    for i, v in enumerate(polymer):
        new_polymer += polymer[i]
        if i < len(products):
            new_polymer += products[i]
    polymer = new_polymer
```

```python

from collections import Counter
import math


def count_polymers(polymer):
    counted = dict(Counter(polymer))
    print(counted)
    min, max = math.inf, 0
    for value in counted.values():
        if value < min:
            min = value
        if value > max:
            max = value

    print(max - min)
```

```python
count_polymers(polymer)
```

```python
pairs = {}
for i in range(len(template)-1):
    pair = template[i:i+2]
    if pair in pairs:
        pairs[pair] += 1
    else:
        pairs[pair] = 1

print(pairs)
steps = 40
for step in range(steps):
    new_pairs = {}
    for pair, v in pairs.items():
        new_letter = instructions_dict[pair]
        pair_0 = pair[0] + new_letter
        pair_1 = new_letter + pair[1]
        
        if pair_0 in new_pairs:
            new_pairs[pair_0] += v
        else:
            new_pairs[pair_0] = v
            
        if pair_1 in new_pairs:
            new_pairs[pair_1] += v
        else:
            new_pairs[pair_1] = v
    print(new_pairs)
    pairs = new_pairs
    
counted = {}
for pair, value in pairs.items():
    for letter in pair:
        if letter in counted:
            counted[letter] += value
        else:
            counted[letter] = value
print(counted)

min, max = math.inf, 0
for value in counted.values():
    if value < min:
        min = value
    if value > max:
        max = value

print(math.floor((max - min) / 2) )
```
